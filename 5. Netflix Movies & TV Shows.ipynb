## Data Wrangling

```python
df.country.fillna("Country Unavailable", inplace=True)
df.director.fillna("Director Unavailable", inplace=True)
df.cast.fillna("Cast Unavailable", inplace = True)
```

```python
df['rating'].unique()
```

```python
df[df['rating'].isnull()]
```

```python
#Replacing missing values of ratings after referring to the internet

rating_replacements = {67: 'TV-PG', 2359: 'TV-14', 3660: 'TV-MA', 3736: 'TV-MA', 3737: 'TV-MA', 3738: 'TV-MA', 4323: 'PG-13'}
for index, rating in rating_replacements.items():
    df.loc[index, 'rating'] = rating
df['rating'].isnull().any()
df.head()
```

```python
df.isnull().any()
```

```python
#Dropped dropped missing values of date added column
df.dropna(subset = ['date_added'], axis = 0, inplace = True)
```

```python
df.isnull().any()
```

```python
df.columns
```

```python
df['date_added'] = pd.to_datetime(df['date_added'])

# Extract month and day name and hour from Start Time after convert
df['month_added'] = df['date_added'].dt.day_name()
df['day_added'] = df['date_added'].dt.day_name()
df['year_added'] = df['date_added'].dt.year
```

## Exploratory Data Analysis

```python
plt.style.use('ggplot')

plt.figure(figsize = (15,18))
plt.subplot(2,2,1)
plt.title('Percentage of Movies vs TV shows', fontsize = 20)
df['type'].value_counts().plot.pie(autopct = "%.2f%%")

plt.subplot(2,2,2)
plt.title('Number of Movies vs TV shows', fontsize = 20)
ax = sns.countplot(x="type",data=df)
for p in ax.patches:
    ax.text(p.get_x() + p.get_width()/2., p.get_height(), '%d' % int(p.get_height()), fontsize=12, color='black', ha='center', va='bottom')
plt.show()
```

#### There are more movies than TV shows on Netflix.

```python
ratings = df['rating'].value_counts()
plt.figure(figsize = (15,8))
ax = sns.countplot(x = 'rating', data = df, palette = 'brg')
total = float(len(df))
for p in ax.patches:
    height = p.get_height()
    ax.text(p.get_x()+p.get_width()/2.,height + 3, '{:.2%}'.format(height/total), ha="center") 
plt.title('Distribution of Ratings', fontsize = 25)
plt.ylabel('Number of Ratings')
plt.xlabel('Rating')
```

```python
movies = df[df['type'] == 'Movie']
tvshows = df[df['type'] == 'TV Show']
```

```python
releases = df[df['release_year']>=2011]
plt.figure(figsize = (15,10))
ax = sns.countplot(x = 'release_year', data = releases, hue = 'type', palette = 'inferno')
plt.title('Number of movies/tv shows released per year on Netflix', fontsize = 20)
ax.set_ylim([0,800])

plt.xlabel('Released Year')
plt.ylabel('Count')
for p in ax.patches:
    ax.text(p.get_x() + p.get_width()/2., p.get_height(), '%d'% (p.get_height()), fontsize = 12, ha = "center", va = "bottom")
    

```

```python
plt.figure(figsize=(10,7))
color = plt.cm.plasma(np.linspace(0,5,25))
ax = tvshows['duration'].value_counts().plot.bar(color = color, figsize = (16,8))
plt.xlabel('Number of Seasons')
plt.ylabel('Count')

for p in ax.patches:
    ax.text(p.get_x() + p.get_width()/2., p.get_height(), '%d'% (p.get_height()), fontsize = 12, ha = "center", va = "bottom")

plt.title('TV Shows Seasons', fontsize = 25)
```

```python
movies['duration'].value_counts().plot.line(figsize = (15,8))
plt.xlabel('Duration of movies')
plt.ylabel('Number of movies')
plt.title('Duration of movies', fontsize = 25)
```

```python
countries = df['country'].apply(lambda x: x.split(",")[0])
countries = countries[countries != 'Country Unavailable']
```

```python
plt.figure(figsize = (15,12))
ax = sns.countplot(y = countries, hue = df['type'], order = countries.value_counts().index [:15])
plt.title('Top 15 countries', fontsize = 25)
for p in ax.patches:
    x_value = p.get_width()
    y_value = p.get_y() + p.get_height() / 2
    space = 20
    ha = 'right'
    label = "{:.0f}".format(x_value)
    plt.annotate(label,(x_value, y_value),xytext=(space, 0),textcoords="offset points", va='center',ha=ha) 
```

```python
top_directors = movies.groupby('director', as_index = False).count()[['director', 'show_id']].sort_values(by = 'show_id', ascending = False)[:11]
plt.figure(figsize = (15,8))
top_directors = top_directors[top_directors != 'Director Unavailable']
ax = sns.barplot(x = top_directors['director'], y = top_directors['show_id'], palette ='Set1_r')
plt.title('Top 10 directors', fontsize  = 25)
plt.ylabel('Count')
plt.xlabel('Director')

for p in ax.patches:
    ax.text(p.get_x() + p.get_width()/2., p.get_height(), '%d'%(p.get_height()), fontsize = 12, va = 'bottom', ha = 'center')
```

```python
genre = df['listed_in'].apply(lambda x: x.split(",")[0])
plt.figure(figsize = (15,14))
ax = sns.countplot(y = genre, order= genre.value_counts().index, palette = 'rainbow')
plt.title('Genres available on Netflix', fontsize = 25)
plt.xlabel('Count')
plt.ylabel('Genres')

for p in ax.patches:
    x_value = p.get_width()
    y_value = p.get_y() + p.get_height()/2
    space = 24
    ha = 'right'
    label = "{:.0f}".format(x_value)
    plt.annotate(label,(x_value, y_value),xytext=(space, 0),textcoords="offset points", va='center',ha=ha)
```

```python
plt.figure(figsize = (15,10))
content_years = df.release_year.value_counts().index[:20]
ax = sns.countplot(x='release_year', data = df, order = content_years, palette = 'copper')
plt.title('Content Released per year on Netflix', fontsize = 25)
plt.xlabel('Year')
plt.ylabel('Count')

for p in ax.patches:
    ax.text(p.get_x() + p.get_width()/2., p.get_height(), '%d'%(p.get_height()), fontsize = 12, va = 'bottom', ha = 'center')

```
