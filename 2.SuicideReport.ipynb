```python
#PIE CHART DESCRIBING THE NO OF SUICIDES FROM EACH GENERATION
plt.style.use('bmh')

df['Generation'].value_counts().plot.pie(figsize=(15,8), explode=(0,0,0,0,0,0), autopct='%1.1f%%')
plt.title('Worldwide Suicide by Generation', fontsize=30)

plt.show()
```

```python
#COUTNPLOT DESCRIBING NO OF SUICIDES IN EACH COUNTRY

plt.style.use('dark_background')
alpha = 0.7
plt.figure(figsize = (10,25))
sns.countplot(y = df['Country'], palette='rainbow')
plt.title('Data by Country', fontsize=30)
plt.show()
```

```python
#HEATMAP DISPLAYING THE CORRELATION BETWEEN DIFFERENT FEATURES

plt.figure(figsize=(20,10))
cor = sns.heatmap(df.corr(), annot = True)
```

```python
#NORMAL BAR GRAPH DESCRIBING THE DISTRIBUTION OF SUICIDE BETWEEN MEN AND WOMEN

plt.figure(figsize=(16,7))
##Plot the graph
sex = sns.countplot(x='Gender',data = df)
plt.title('Suicides per Gender', fontsize=30)
```

```python
plt.figure(figsize=(16,7))
plt.style.use('dark_background')

bar_age = sns.barplot(x = 'Gender', y = 'NoOfSuicides', hue = 'Age',data = df, palette="Paired")
plt.title('No of Suicides Per Age Group divided by Gender', fontsize = 30)
```

```python
plt.figure(figsize=(15,8))

bar_gen = sns.barplot(x = 'Gender', y = 'NoOfSuicides', hue = 'Generation', data=df, palette = 'rainbow')
plt.title('No of Suicides for each Generation divided by Gender', fontsize = 30)
```

```python
cat_accord_year = sns.catplot('Gender','NoOfSuicides',hue='Age',col='Year',data=df,kind='bar',col_wrap=4, palette = 'bright')
plt.title('Suicides per Year divided by Gender and Age ', fontsize = 20)
```

```python
df['Age']
```

```python
age_5 = df.loc[df.loc[:, 'Age']=='5-14 years',:]
age_15 = df.loc[df.loc[:, 'Age']=='15-24 years',:]
age_25 = df.loc[df.loc[:, 'Age']=='25-34 years',:]
age_35 = df.loc[df.loc[:, 'Age']=='35-54 years',:]
age_55 = df.loc[df.loc[:, 'Age']== '55-74 years', :]
age_75 = df.loc[df.loc[:, 'Age']=='75+', :]

plt.figure(figsize=(16,8))
plt.style.use('bmh')
plt.title('LinePlot over the years for each age group', fontsize = 25)

age_5_lp = sns.lineplot(x = 'Year', y = 'NoOfSuicides', data = age_5)
age_15_lp = sns.lineplot(x = 'Year', y = 'NoOfSuicides', data = age_15)
age_25_lp = sns.lineplot(x = 'Year', y = 'NoOfSuicides', data = age_25)
age_35_lp = sns.lineplot(x = 'Year', y = 'NoOfSuicides', data = age_35)
age_55_lp = sns.lineplot(x = 'Year', y = 'NoOfSuicides', data = age_55)
age_75_lp = sns.lineplot(x = 'Year', y = 'NoOfSuicides', data = age_75)

leg = plt.legend(['5-14 years', '15-24 years', '25-34 years', '35-54 years', '55-74 years', '75+ years'])
```

```python
malepop = df.loc[df.loc[:, 'Gender'] == 'male',:]
femalepop = df.loc[df.loc[:, 'Gender']== 'female']

plt.style.use('bmh')

plt.figure(figsize = (16,7))
lpmale = sns.lineplot(x = 'Year', y = 'NoOfSuicides', data = malepop)
lpfemale = sns.lineplot(x = 'Year',y = 'NoOfSuicides', data = femalepop)
leg1 = plt.legend(['Male', 'Female'])
plt.title('No of Suicides over the years w.r.t Gender', fontsize = 20)
```

```python

f,ax1 = plt.subplots(figsize =(20,10))
plt.style.use('ggplot')
sns.pointplot(x = 'Year', y = 'NoOfSuicides', data = df, alpha = 0.8)
plt.xticks(rotation = 40)

plt.title('Year vs Suicide Values', fontsize = 20)
plt.grid()

```
