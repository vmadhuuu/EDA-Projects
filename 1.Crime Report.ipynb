```python
#now we need to fill the missing value with the MODE of that column

df['PdDistrict'].fillna(df['PdDistrict'].mode()[0], inplace=True)
df.isnull().any()
```

```python
df
```

```python
#so now im making a plot which counts how many cases are there for each category
#here the params is used to change the size of the plot
#the style used is the black background
#palette is the type of color style used for a visaully pleasing plot


plt.rcParams['figure.figsize']=(25,8)
plt.style.use('dark_background')


sns.countplot(df['Category'], palette = 'gnuplot')
plt.title('Major Crimes in Sanfrancisco', fontweight=30, fontsize=40)
plt.xticks(rotation=90)
plt.show()
```

```python
#this is another plot for the same purpose
#this keeps category in the y axis instead



plt.figure(figsize=(15,10))
plt.title('Major Crimes in Sanfrancisco')

sns.countplot(y=df['Category'])


plt.show()
```

```python
#CREATING A TREEMAP
#so were taking the top 25 values for showing the no of cases for each category
#so fivethirtyeight is the name of the style which is used here
y = df['Category'].value_counts().head(25)
plt.rcParams['figure.figsize'] = (15, 15)
plt.style.use('fivethirtyeight')

#now the cm stands for CMAP
#viridis can be changed to whatever style you want to use
#linspace(...15) means the degree of darkness of the colors
#sizes is denoting on what basis the squares should be enlarged
#and the labels denote which data should be used for denoting the squares


color = plt.cm.viridis(np.linspace(0,1,15))
squarify.plot(sizes = y.values, label = y.index, alpha=.8, color = color)
plt.title('Tree Map for Top 25 Crimes', fontsize = 20)

plt.axis('off')
plt.show()
```

```python
#now this is a different type of viz for the purpose: desription of the crime

from wordcloud import WordCloud
plt.rcParams['figure.figsize']=(15,15)
plt.style.use('classic')


wc = WordCloud(background_color = 'black', width = 1500, height = 1500).generate(str(df['Descript']))
plt.title('Description of Crime', fontsize = 20)

#imshow creates an image from an array
plt.imshow(wc)
plt.axis('off')
plt.show()
```

```python
plt.rcParams['figure.figsize'] = (10,10)
plt.style.use('dark_background')

color = plt.cm.winter(np.linspace(0,1,15))
df['PdDistrict'].value_counts().plot.bar(color = color, figsize = (10,10))
plt.title('District with most Crime',fontsize = 30)
plt.xticks(rotation = 90)
plt.show()
```

```python
plt.rcParams['figure.figsize'] = (10,10)
plt.style.use('dark_background')
color = plt.cm.autumn(np.linspace(0, 1, 15))
df['Address'].value_counts().head(15).plot.bar(color=color, figsize = (15,8))
plt.title('Top 15 regions in crime', fontsize=30)

plt.xticks(rotation=90)
plt.show()
```

```python
#CREATING A PIE CHART

plt.style.use('bmh')

df['DayOfWeek'].value_counts().head(15).plot.pie(figsize=(15, 8), explode=(0, 0, 0, 0, 0, 0, 0))
plt.title('Crime Count per Day', fontsize = 20)

plt.xticks(rotation=90)
plt.show()
```

```python

plt.style.use('seaborn')

color = plt.cm.winter(np.linspace(0,14,25))
df['Resolution'].value_counts().plot.bar(color=color, figsize=(15,8))
plt.title('Resolutions for Crime', fontsize = 30)

plt.xticks(rotation=90)
plt.show()
```

```python
#converting the date values to a certain datetime format
df['Date'] = pd.to_datetime(df['Date'])

#extracting the month out of the datetime formatted date values; creating a new seperate column for that
df['Month'] = df['Date'].dt.month

plt.style.use('fivethirtyeight')
plt.rcParams['figure.figsize'] = (15,8)

sns.countplot(df['Month'], palette = 'autumn')
plt.title('Crimes per Month', fontsize=30)
plt.show()
```

```python
import warnings
warnings.filterwarnings('ignore')

color = plt.cm.twilight(np.linspace(0,5,100))
df['Time'].value_counts().head(20).plot.bar(color = color, figsize =(15,9))
plt.title('Distribution of Crime over day', fontsize=30)
plt.show()
```
